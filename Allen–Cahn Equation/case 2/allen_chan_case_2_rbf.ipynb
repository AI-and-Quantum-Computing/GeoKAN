```python
import os
import math
import time
import random
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

from efficientkan import KAN_twovariable as KAN_2

# ============================================================
# User controls (single-line device selection)
# ============================================================
DEVICE ="cpu" # "cuda:0" if torch.cuda.is_available() else "cpu"   # change to "cpu" manually if needed
DTYPE = torch.float32                                          # keep PIKAN dtype aligned with notebook
LMKAN_DTYPE = torch.float64                                    # LM-KAN benefits from float64 for 2nd derivatives
SEED = 0
OUTPUT_DIR = Path("./output_rbf")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# PIKAN: kept exactly as in uploaded training notebook
PIKAN_LAYERS = [2, 8, 6, 1]
PIKAN_LR = 1e-3
PIKAN_EPOCHS = 20000
PIKAN_V = 1e-4
PIKAN_DATA_FRACTION = 0.10

# LM-KAN: adjusted ONLY to make parameter count comparable to PIKAN
# This configuration gives ~700 trainable parameters.
LMKAN_WIDTH = 5
LMKAN_DEPTH = 2
LMKAN_K = 7
LMKAN_METRIC_HIDDEN = 9
LMKAN_GAMMA = 2.0
LMKAN_BASIS = "rbf"     # choose "wavelet" or "rbf"
LMKAN_LR = 2e-3
LMKAN_STEPS = 20000
LMKAN_NR = 8000
LMKAN_NIC = 512
LMKAN_NBC = 512
LMKAN_WDATA = 5.0

# Numerical grid: kept exactly as in notebook
Nx = 100
Nt = 100
L = 2
T_total = 1
alpha = 0.0001
gamma = 5

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 16
plt.rcParams['font.weight'] = 'bold'


def seed_all(seed=0):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)


seed_all(SEED)


def get_torch_device(device_str):
    if device_str.startswith("cuda") and not torch.cuda.is_available():
        return torch.device("cpu")
    return torch.device(device_str)


device = get_torch_device(DEVICE)
print(f"Using device: {device}")


def count_trainable_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)


# ============================================================
# 1) Numerical reference (kept consistent with uploaded notebook)
# ============================================================

def solve_allen_cahn_case2_notebook_style(Nx=100, Nt=100, alpha=1e-4, gamma=5.0):
    dx = L / (Nx - 1)
    dt = T_total / (Nt - 1)

    x = torch.linspace(-1, 1, Nx, dtype=torch.float32)
    t_vec = torch.linspace(0, 1, Nt, dtype=torch.float32)

    u = torch.zeros((Nt, Nx), dtype=torch.float32)
    u[0, :] = x ** 2 * torch.cos(np.pi * x)

    # notebook-style periodic-like boundary handling
    u[:, 0] = u[:, -1]

    for n in range(0, Nt - 1):
        for i in range(1, Nx - 1):
            uxx = (u[n, i + 1] - 2 * u[n, i] + u[n, i - 1]) / dx ** 2
            u_t = alpha * uxx - gamma * (u[n, i] ** 3 - u[n, i])
            u[n + 1, i] = u[n, i] + dt * u_t

        u[n + 1, 0] = u[n + 1, -1]
        u[n + 1, -1] = u[n + 1, 0]

    return x, t_vec, u


# ============================================================
# 2) PIKAN loss (kept exactly as notebook logic)
# ============================================================

def pinn_loss_pikan(model, x, t, x_ic, t_ic, u_ic, u, v=0.0001):
    dev = x.device
    u_pred = model(x, t)
    zeros = torch.zeros_like(t).to(dev)
    ones = torch.ones_like(t).to(dev)

    t_s = torch.linspace(0, 1, 100, device=dev)
    num_samples = int(0.1 * len(t_s))
    indices = torch.randperm(len(t_s), device=dev)
    selected_indices = indices[:num_samples]
    t_sharp = t_s[selected_indices].to(dev)

    u_t = torch.autograd.grad(
        u_pred, t, grad_outputs=torch.ones_like(u_pred).to(dev), create_graph=True, retain_graph=True
    )[0]
    u_x = torch.autograd.grad(
        u_pred, x, grad_outputs=torch.ones_like(u_pred).to(dev), create_graph=True, retain_graph=True
    )[0]
    u_xx = torch.autograd.grad(
        u_x, x, grad_outputs=torch.ones_like(u_x).to(dev), create_graph=True, retain_graph=True
    )[0]

    f_pred = u_t - v * u_xx + 5 * u_pred ** 3 - 5 * u_pred

    ic_pred = model(x_ic, t_ic)
    ic_loss = torch.mean((ic_pred - u_ic) ** 2)

    x_minus_ones = -1 * torch.ones_like(t).to(dev).requires_grad_(True)
    x_plus_ones = torch.ones_like(t).to(dev).requires_grad_(True)

    u_minus_ones = model(x_minus_ones, t)
    u_plus_ones = model(x_plus_ones, t)

    u_x_minus_ones = torch.autograd.grad(
        u_minus_ones, x_minus_ones, grad_outputs=torch.ones_like(u_minus_ones).to(dev), create_graph=True, retain_graph=True
    )[0]
    u_x_plus_ones = torch.autograd.grad(
        u_plus_ones, x_plus_ones, grad_outputs=torch.ones_like(u_plus_ones).to(dev), create_graph=True, retain_graph=True
    )[0]

    bc_loss = torch.mean((u_minus_ones - u_plus_ones) ** 2) + torch.mean((u_x_minus_ones - u_x_plus_ones) ** 2)

    loss_sharp = 0
    for t_eval in t_sharp:
        t_knot = t_eval * ones
        u_sharp = model(x_ic, t_knot)
        t_index = torch.argmin(torch.abs(t - t_eval))
        u_numerical_sharp = u[t_index, :].view(-1, 1).to(dev).requires_grad_(True)
        loss_sharp = loss_sharp + (u_numerical_sharp - u_sharp) ** 2

    loss = torch.mean(f_pred ** 2 + loss_sharp) + ic_loss + bc_loss

    parts = {
        "residual": torch.mean(f_pred ** 2).detach(),
        "sharp": torch.mean(loss_sharp).detach() if torch.is_tensor(loss_sharp) else torch.tensor(float(loss_sharp), device=dev),
        "ic": ic_loss.detach(),
        "bc": bc_loss.detach(),
        "total": loss.detach(),
    }
    return loss, parts


# ============================================================
# 3) LM-KAN model (same model, only smaller config)
# ============================================================
class MetricNet(nn.Module):
    def __init__(self, d_in, hidden=64, eps=1e-8):
        super().__init__()
        self.eps = eps
        self.net = nn.Sequential(
            nn.Linear(d_in, hidden), nn.SiLU(),
            nn.Linear(hidden, hidden), nn.SiLU(),
            nn.Linear(hidden, d_in),
        )

    def forward(self, x):
        return F.softplus(self.net(x)) + self.eps


def mexican_hat(u):
    return (1.0 - u * u) * torch.exp(-0.5 * u * u)


def rbf_gamma(u, gamma):
    return torch.exp(-gamma * (u * u))


class LearnedMetricBasisLayer(nn.Module):
    def __init__(self, in_dim, out_dim, K=16, basis="wavelet", gamma=2.0, metric_hidden=64):
        super().__init__()
        assert basis in ["wavelet", "rbf"]
        self.in_dim = in_dim
        self.K = K
        self.basis = basis
        self.gamma = float(gamma)
        self.metric = MetricNet(in_dim, hidden=metric_hidden)

        grid = torch.linspace(-2.5, 2.5, K).view(1, K).repeat(in_dim, 1)
        self.centers = nn.Parameter(grid + 0.05 * torch.randn(in_dim, K))
        self.log_widths = nn.Parameter(torch.log(0.8 * torch.ones(in_dim, K)))
        self.out = nn.Linear(in_dim * K + 1, out_dim)

    def forward(self, x):
        g = self.metric(x)
        z = x * torch.sqrt(g)
        widths = torch.exp(self.log_widths).unsqueeze(0) + 1e-12
        u = (z.unsqueeze(-1) - self.centers.unsqueeze(0)) / widths

        if self.basis == "wavelet":
            phi = mexican_hat(u)
        else:
            phi = rbf_gamma(u, gamma=self.gamma)

        phi = phi.reshape(x.shape[0], -1)
        logdet = torch.log(g).sum(dim=1, keepdim=True)
        feats = torch.cat([phi, logdet], dim=1)
        return self.out(feats)


class LMKAN(nn.Module):
    def __init__(self, in_dim, width=32, depth=2, K=16, basis="wavelet", gamma=2.0, metric_hidden=64):
        super().__init__()
        layers = []
        d = in_dim
        for _ in range(depth):
            layers.append(LearnedMetricBasisLayer(d, width, K=K, basis=basis, gamma=gamma, metric_hidden=metric_hidden))
            layers.append(nn.Tanh())
            d = width
        self.body = nn.Sequential(*layers)
        self.head = nn.Linear(d, 1)

    def forward(self, x):
        return self.head(self.body(x))


# ============================================================
# 4) LM-KAN PDE utilities
# ============================================================

def sample_uniform_2d(n, x_lo, x_hi, t_lo, t_hi):
    x = np.random.uniform(x_lo, x_hi, size=(n, 1))
    t = np.random.uniform(t_lo, t_hi, size=(n, 1))
    return np.hstack([x, t])


def gradients(u, x):
    return torch.autograd.grad(u, x, grad_outputs=torch.ones_like(u), create_graph=True)[0]


def second_derivative(u, x, dim):
    du = gradients(u, x)[:, dim:dim + 1]
    d2u = torch.autograd.grad(du, x, grad_outputs=torch.ones_like(du), create_graph=True)[0][:, dim:dim + 1]
    return d2u


def to_torch(a, dtype, device):
    return torch.tensor(a, device=device, dtype=dtype)


def make_data_points_from_grid(x, t, U, frac=0.10, seed=0):
    rng = np.random.RandomState(seed)
    x_np = np.asarray(x)
    t_np = np.asarray(t)
    U_np = np.asarray(U)
    Nt, Nx = U_np.shape
    total = Nt * Nx
    m = max(1, int(frac * total)) if frac > 0 else 0
    idx = rng.choice(total, size=m, replace=False)
    it = idx // Nx
    ix = idx % Nx
    Xd = np.stack([x_np[ix], t_np[it]], axis=1)
    Ud = U_np[it, ix].reshape(-1, 1)
    return Xd, Ud


def loss_allen_cahn_lmkan(model, nu, data_pts=None, w_data=1.0, n_r=5000, n_ic=512, n_bc=512, dtype=torch.float64, device=torch.device('cpu')):
    Xr = to_torch(sample_uniform_2d(n_r, -1.0, 1.0, 0.0, 1.0), dtype=dtype, device=device)
    Xr.requires_grad_(True)
    u = model(Xr)
    grads = gradients(u, Xr)
    u_t = grads[:, 1:2]
    u_xx = second_derivative(u, Xr, dim=0)
    r = u_t - nu * u_xx + 5 * (u ** 3) - 5 * u
    Lr = (r ** 2).mean()

    x_ic = np.random.uniform(-1.0, 1.0, size=(n_ic, 1))
    u0 = (x_ic ** 2) * np.cos(np.pi * x_ic)
    Xic = to_torch(np.hstack([x_ic, np.zeros((n_ic, 1))]), dtype=dtype, device=device)
    Lic = ((model(Xic) - to_torch(u0, dtype=dtype, device=device)) ** 2).mean()

    t_bc = np.random.uniform(0.0, 1.0, size=(n_bc, 1))
    XbL = to_torch(np.hstack([-np.ones((n_bc, 1)), t_bc]), dtype=dtype, device=device).requires_grad_(True)
    XbR = to_torch(np.hstack([ np.ones((n_bc, 1)), t_bc]), dtype=dtype, device=device).requires_grad_(True)
    uL = model(XbL)
    uR = model(XbR)
    uL_x = gradients(uL, XbL)[:, 0:1]
    uR_x = gradients(uR, XbR)[:, 0:1]
    Lbc = ((uL - uR) ** 2).mean() + ((uL_x - uR_x) ** 2).mean()

    Ldata = torch.tensor(0.0, dtype=dtype, device=device)
    if data_pts is not None:
        Xd, Ud = data_pts
        pred = model(Xd)
        Ldata = ((pred - Ud) ** 2).mean()

    total = Lr + Lic + Lbc + w_data * Ldata
    parts = {
        "residual": Lr.detach(),
        "ic": Lic.detach(),
        "bc": Lbc.detach(),
        "data": Ldata.detach(),
        "total": total.detach(),
    }
    return total, parts


# ============================================================
# 5) Training / inference helpers
# ============================================================

def maybe_wrap_dataparallel(model, device):
    if device.type == "cuda" and torch.cuda.device_count() > 1:
        return nn.DataParallel(model)
    return model


def unwrap_state_dict(model):
    return model.module.state_dict() if isinstance(model, nn.DataParallel) else model.state_dict()


def train_pikan(model, x, t, x_ic, t_ic, u_ic, u_ref, num_epochs=20000, lr=1e-3, v=1e-4):
    optimizer = optim.AdamW(model.parameters(), lr=lr)
    min_loss = float('inf')
    best_model_state = None
    history = {
        "epoch": [], "loss": [], "residual": [], "sharp": [], "ic": [], "bc": []
    }

    start_time = time.time()
    for epoch in range(num_epochs):
        optimizer.zero_grad()
        loss, parts = pinn_loss_pikan(model, x, t, x_ic, t_ic, u_ic, u_ref, v=v)
        loss.backward(retain_graph=True)
        optimizer.step()

        history["epoch"].append(epoch)
        history["loss"].append(float(loss.detach().cpu()))
        history["residual"].append(float(parts["residual"].cpu()))
        history["sharp"].append(float(parts["sharp"].mean().cpu()))
        history["ic"].append(float(parts["ic"].cpu()))
        history["bc"].append(float(parts["bc"].cpu()))

        current = float(loss.detach().cpu())
        if current < min_loss:
            min_loss = current
            best_model_state = {k: v.detach().cpu().clone() for k, v in unwrap_state_dict(model).items()}

        if epoch % 1000 == 0:
            print(f"[PIKAN] Epoch {epoch:5d} | Loss={current:.6e} | Residual={history['residual'][-1]:.6e} | Sharp={history['sharp'][-1]:.6e}")

    elapsed = time.time() - start_time
    return best_model_state, history, elapsed, min_loss


def train_lmkan(model, loss_builder, steps=12000, lr=2e-3, clip=1.0):
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-6)
    best = float("inf")
    best_sd = None
    history = {
        "step": [], "loss": [], "residual": [], "ic": [], "bc": [], "data": []
    }

    start_time = time.time()
    for it in range(1, steps + 1):
        optimizer.zero_grad(set_to_none=True)
        loss, parts = loss_builder()
        loss.backward()
        if clip is not None:
            nn.utils.clip_grad_norm_(model.parameters(), clip)
        optimizer.step()

        cur = float(loss.detach().cpu())
        history["step"].append(it)
        history["loss"].append(cur)
        history["residual"].append(float(parts["residual"].cpu()))
        history["ic"].append(float(parts["ic"].cpu()))
        history["bc"].append(float(parts["bc"].cpu()))
        history["data"].append(float(parts["data"].cpu()))

        if cur < best:
            best = cur
            best_sd = {k: v.detach().cpu().clone() for k, v in model.state_dict().items()}

        if it % 600 == 0:
            print(f"[LM-KAN] Step {it:5d} | Loss={cur:.6e} | Residual={history['residual'][-1]:.6e} | Data={history['data'][-1]:.6e}")

    elapsed = time.time() - start_time
    return best_sd, history, elapsed, best


@torch.no_grad()
def predict_pikan_grid(model, x_vals, t_vals, device):
    model.eval()
    preds = np.zeros((len(t_vals), len(x_vals)), dtype=np.float32)
    x_tensor = torch.tensor(x_vals, dtype=torch.float32, device=device).view(-1, 1)
    for j, tv in enumerate(t_vals):
        t_tensor = (float(tv) * torch.ones_like(x_tensor)).to(device)
        preds[j] = model(x_tensor, t_tensor).detach().cpu().numpy().reshape(-1)
    return preds


@torch.no_grad()
def predict_lmkan_grid(model, x_vals, t_vals, device, dtype=torch.float64, chunk=8192):
    model.eval()
    Xg = np.stack([np.tile(x_vals, len(t_vals)), np.repeat(t_vals, len(x_vals))], axis=1)
    Xg_t = torch.tensor(Xg, device=device, dtype=dtype)
    preds = []
    for i in range(0, Xg_t.shape[0], chunk):
        y = model(Xg_t[i:i + chunk]).detach().cpu().numpy()
        preds.append(y)
    return np.vstack(preds).reshape(len(t_vals), len(x_vals)).astype(np.float32)


def compute_residual_grid_pikan(model, x_vals, t_vals, device, v=1e-4):
    model.eval()
    residual = np.zeros((len(t_vals), len(x_vals)), dtype=np.float32)
    for j, tv in enumerate(t_vals):
        x_tensor = torch.tensor(x_vals, dtype=torch.float32, device=device).view(-1, 1).requires_grad_(True)
        t_tensor = (float(tv) * torch.ones_like(x_tensor)).to(device).requires_grad_(True)
        u_pred = model(x_tensor, t_tensor)
        u_t = torch.autograd.grad(u_pred, t_tensor, grad_outputs=torch.ones_like(u_pred), create_graph=False, retain_graph=True)[0]
        u_x = torch.autograd.grad(u_pred, x_tensor, grad_outputs=torch.ones_like(u_pred), create_graph=True, retain_graph=True)[0]
        u_xx = torch.autograd.grad(u_x, x_tensor, grad_outputs=torch.ones_like(u_x), create_graph=False, retain_graph=False)[0]
        f = u_t - v * u_xx + 5 * u_pred ** 3 - 5 * u_pred
        residual[j] = f.detach().cpu().numpy().reshape(-1)
    return residual


def compute_residual_grid_lmkan(model, x_vals, t_vals, device, v=1e-4, dtype=torch.float64):
    model.eval()
    residual = np.zeros((len(t_vals), len(x_vals)), dtype=np.float32)
    for j, tv in enumerate(t_vals):
        x_tensor = torch.tensor(x_vals, dtype=dtype, device=device).view(-1, 1)
        t_tensor = (float(tv) * torch.ones_like(x_tensor)).to(device)
        X = torch.cat([x_tensor, t_tensor], dim=1).requires_grad_(True)
        u_pred = model(X)
        grads = torch.autograd.grad(u_pred, X, grad_outputs=torch.ones_like(u_pred), create_graph=True)[0]
        u_t = grads[:, 1:2]
        u_x = grads[:, 0:1]
        u_xx = torch.autograd.grad(u_x, X, grad_outputs=torch.ones_like(u_x), create_graph=False)[0][:, 0:1]
        f = u_t - v * u_xx + 5 * u_pred ** 3 - 5 * u_pred
        residual[j] = f.detach().cpu().numpy().reshape(-1)
    return residual


def grid_metrics(U_ref, U_pred):
    err = U_pred - U_ref
    mse = float(np.mean(err ** 2))
    mae = float(np.mean(np.abs(err)))
    rel_l2 = float(np.linalg.norm(err.ravel()) / (np.linalg.norm(U_ref.ravel()) + 1e-12))
    return {"mse": mse, "mae": mae, "rel_l2": rel_l2}


def save_checkpoint(path, model_name, model_state_dict, config, history, metrics):
    torch.save({
        "model_name": model_name,
        "model_state_dict": model_state_dict,
        "config": config,
        "history": history,
        "metrics": metrics,
    }, path)


def plot_loss_comparison(pikan_history, lmkan_history, save_path):
    plt.figure(figsize=(9, 6), dpi=300)
    plt.plot(pikan_history["epoch"], np.log(np.asarray(pikan_history["loss"]) + 1e-12), linewidth=2.5, label="PIKAN")
    plt.plot(lmkan_history["step"], np.log(np.asarray(lmkan_history["loss"]) + 1e-12), linewidth=2.5, label=f"LM-KAN ({LMKAN_BASIS})")
    plt.xlabel("Iterations", fontsize=18, weight="bold")
    plt.ylabel("Log(Loss)", fontsize=18, weight="bold")
    plt.legend(fontsize=16)
    plt.grid(False)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()


def plot_component_losses(pikan_history, lmkan_history, save_path):
    fig, axes = plt.subplots(1, 2, figsize=(16, 5), dpi=300)

    axes[0].plot(pikan_history["epoch"], np.log(np.asarray(pikan_history["residual"]) + 1e-12), label="Residual", linewidth=2.0)
    axes[0].plot(pikan_history["epoch"], np.log(np.asarray(pikan_history["sharp"]) + 1e-12), label="Sharp/Data", linewidth=2.0)
    axes[0].plot(pikan_history["epoch"], np.log(np.asarray(pikan_history["ic"]) + 1e-12), label="IC", linewidth=2.0)
    axes[0].plot(pikan_history["epoch"], np.log(np.asarray(pikan_history["bc"]) + 1e-12), label="BC", linewidth=2.0)
    axes[0].set_title("PIKAN loss parts", fontsize=18, weight="bold")
    axes[0].set_xlabel("Epoch")
    axes[0].set_ylabel("Log(loss)")
    axes[0].legend(fontsize=11)
    axes[0].grid(False)

    axes[1].plot(lmkan_history["step"], np.log(np.asarray(lmkan_history["residual"]) + 1e-12), label="Residual", linewidth=2.0)
    axes[1].plot(lmkan_history["step"], np.log(np.asarray(lmkan_history["data"]) + 1e-12), label="Data", linewidth=2.0)
    axes[1].plot(lmkan_history["step"], np.log(np.asarray(lmkan_history["ic"]) + 1e-12), label="IC", linewidth=2.0)
    axes[1].plot(lmkan_history["step"], np.log(np.asarray(lmkan_history["bc"]) + 1e-12), label="BC", linewidth=2.0)
    axes[1].set_title(f"LM-KAN ({LMKAN_BASIS}) loss parts", fontsize=18, weight="bold")
    axes[1].set_xlabel("Step")
    axes[1].set_ylabel("Log(loss)")
    axes[1].legend(fontsize=11)
    axes[1].grid(False)

    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()


def plot_time_slices(x_vals, t_vals, U_num, U_pikan, U_lmkan, save_path_prefix):
    times = [0.0, 0.05, 0.9]
    for t_val in times:
        t_index = int(np.argmin(np.abs(t_vals - t_val)))
        plt.figure(figsize=(8, 4.8), dpi=300)
        plt.plot(x_vals, U_pikan[t_index], linestyle='-', linewidth=2.5, color='b', label=f'PIKAN (t={t_vals[t_index]:.2f})')
        plt.plot(x_vals, U_lmkan[t_index], linestyle='-.', linewidth=2.5, color='orange', label=f'LM-KAN (t={t_vals[t_index]:.2f})')
        plt.plot(x_vals, U_num[t_index], linestyle='--', marker='*', markersize=8, color='black', linewidth=2.5,
                 label=f'Numerical (t={t_vals[t_index]:.2f})', markevery=10)
        plt.ylabel('u(x, t)', fontsize=20, weight='bold')
        plt.xlabel('x', fontsize=20, weight='bold')
        plt.legend(fontsize=12, frameon=True)
        plt.grid(True)
        plt.tight_layout()
        plt.savefig(f"{save_path_prefix}_t{t_vals[t_index]:.2f}.png", dpi=300, bbox_inches='tight')
        plt.show()


def plot_solution_comparison(U_num, U_pikan, U_lmkan, x_vals, t_vals, save_path):
    fig, axes = plt.subplots(1, 3, figsize=(21, 6), dpi=300, sharey=True)
    vmin = min(U_num.min(), U_pikan.min(), U_lmkan.min())
    vmax = max(U_num.max(), U_pikan.max(), U_lmkan.max())

    for ax, data, title in zip(axes, [U_num, U_pikan, U_lmkan], ["Numerical", "PIKAN", f"LM-KAN ({LMKAN_BASIS})"]):
        im = ax.imshow(data, extent=[-1, 1, 0, 1], origin='lower', aspect='auto', cmap='viridis', interpolation='none', vmin=vmin, vmax=vmax)
        ax.set_title(title, fontsize=18, weight='bold')
        ax.set_xlabel('x', fontsize=16, weight='bold')
        ax.set_ylabel('t', fontsize=16, weight='bold')
        ax.grid(False)
        cbar = fig.colorbar(im, ax=ax)
        cbar.set_label(r'u(x,t)', fontsize=14, weight='bold')

    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()


def plot_error_and_residual(U_num, U_pikan, U_lmkan, R_pikan, R_lmkan, save_path):
    err_pikan = np.abs(U_pikan - U_num)
    err_lmkan = np.abs(U_lmkan - U_num)
    abs_r_pikan = np.abs(R_pikan)
    abs_r_lmkan = np.abs(R_lmkan)

    fig, axes = plt.subplots(2, 2, figsize=(16, 10), dpi=300, sharex=True, sharey=True)
    panels = [
        (err_pikan, '|PIKAN - Numerical|'),
        (err_lmkan, f'|LM-KAN ({LMKAN_BASIS}) - Numerical|'),
        (abs_r_pikan, '|PIKAN residual|'),
        (abs_r_lmkan, f'|LM-KAN ({LMKAN_BASIS}) residual|'),
    ]

    for ax, (data, title) in zip(axes.ravel(), panels):
        im = ax.imshow(data, extent=[-1, 1, 0, 1], origin='lower', aspect='auto', cmap='magma', interpolation='none')
        ax.set_title(title, fontsize=16, weight='bold')
        ax.set_xlabel('x', fontsize=14, weight='bold')
        ax.set_ylabel('t', fontsize=14, weight='bold')
        ax.grid(False)
        fig.colorbar(im, ax=ax)

    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()


# ============================================================
# 6) Main run
# ============================================================
if __name__ == "__main__":
    x_ref_t, t_ref_t, u_ref_t = solve_allen_cahn_case2_notebook_style(Nx=Nx, Nt=Nt, alpha=alpha, gamma=gamma)
    x_vals = x_ref_t.cpu().numpy().astype(np.float32)
    t_vals = t_ref_t.cpu().numpy().astype(np.float32)
    U_num = u_ref_t.cpu().numpy().astype(np.float32)

    # ---------- PIKAN data ----------
    x = torch.linspace(-1, 1, Nx, dtype=DTYPE).view(-1, 1).requires_grad_(True).to(device)
    t = torch.linspace(0, 1, Nt, dtype=DTYPE).view(-1, 1).requires_grad_(True).to(device)
    x_ic = x.clone().requires_grad_(True)
    t_ic = torch.zeros_like(x_ic).requires_grad_(True)
    u_ic = (x_ic ** 2 * torch.cos(np.pi * x_ic)).view(-1, 1)
    u_ref_dev = u_ref_t.to(device=device, dtype=DTYPE)

    pikan_model = KAN_2(PIKAN_LAYERS)
    pikan_model = maybe_wrap_dataparallel(pikan_model, device)
    pikan_model = pikan_model.to(device)

    # ---------- LM-KAN data ----------
    Xd_np, Ud_np = make_data_points_from_grid(x_vals, t_vals, U_num, frac=PIKAN_DATA_FRACTION, seed=SEED)
    Xd = torch.tensor(Xd_np, dtype=LMKAN_DTYPE, device=device)
    Ud = torch.tensor(Ud_np, dtype=LMKAN_DTYPE, device=device)

    lmkan_model = LMKAN(
        in_dim=2,
        width=LMKAN_WIDTH,
        depth=LMKAN_DEPTH,
        K=LMKAN_K,
        basis=LMKAN_BASIS,
        gamma=LMKAN_GAMMA,
        metric_hidden=LMKAN_METRIC_HIDDEN,
    ).to(device=device, dtype=LMKAN_DTYPE)

    # ---------- Parameter counts ----------
    pikan_param_count = count_trainable_parameters(pikan_model.module if isinstance(pikan_model, nn.DataParallel) else pikan_model)
    lmkan_param_count = count_trainable_parameters(lmkan_model)

    print("\n=== Model parameter counts ===")
    print(f"PIKAN trainable parameters   : {pikan_param_count:,}")
    print(f"LM-KAN trainable parameters : {lmkan_param_count:,}")

    def lmkan_loss_builder():
        return loss_allen_cahn_lmkan(
            lmkan_model,
            nu=alpha,
            data_pts=(Xd, Ud),
            w_data=LMKAN_WDATA,
            n_r=LMKAN_NR,
            n_ic=LMKAN_NIC,
            n_bc=LMKAN_NBC,
            dtype=LMKAN_DTYPE,
            device=device,
        )

    print("\nTraining PIKAN with notebook-exact configuration...")

    from timeit import default_timer as timer
    start_time_pikan = timer()
    pikan_best_sd, pikan_history, pikan_time, pikan_best = train_pikan(
        pikan_model, x, t, x_ic, t_ic, u_ic, u_ref_dev, num_epochs=PIKAN_EPOCHS, lr=PIKAN_LR, v=PIKAN_V
    )
    end_time_pikan = timer()
    print(f"Total training time pikan: {end_time_pikan-start_time_pikan:.3f} seconds")
    pikan_model.load_state_dict(pikan_best_sd)

    print(f"\nTraining LM-KAN ({LMKAN_BASIS}) with parameter-matched configuration...")
    start_time_lmkan = timer()
    lmkan_best_sd, lmkan_history, lmkan_time, lmkan_best = train_lmkan(
        lmkan_model, lmkan_loss_builder, steps=LMKAN_STEPS, lr=LMKAN_LR, clip=1.0
    )
    end_time_lmkan = timer()
    print(f"Total training time lmkan: {end_time_lmkan-start_time_lmkan:.3f} seconds")
    lmkan_model.load_state_dict(lmkan_best_sd)

    # ---------- Predictions ----------
    U_pikan = predict_pikan_grid(pikan_model, x_vals, t_vals, device)
    U_lmkan = predict_lmkan_grid(lmkan_model, x_vals, t_vals, device, dtype=LMKAN_DTYPE)

    R_pikan = compute_residual_grid_pikan(pikan_model, x_vals, t_vals, device, v=alpha)
    R_lmkan = compute_residual_grid_lmkan(lmkan_model, x_vals, t_vals, device, v=alpha, dtype=LMKAN_DTYPE)

    pikan_metrics = grid_metrics(U_num, U_pikan)
    lmkan_metrics = grid_metrics(U_num, U_lmkan)
    pikan_metrics["best_loss"] = pikan_best
    pikan_metrics["train_time_sec"] = pikan_time
    pikan_metrics["residual_mse"] = float(np.mean(R_pikan ** 2))
    pikan_metrics["num_parameters"] = int(pikan_param_count)

    lmkan_metrics["best_loss"] = lmkan_best
    lmkan_metrics["train_time_sec"] = lmkan_time
    lmkan_metrics["residual_mse"] = float(np.mean(R_lmkan ** 2))
    lmkan_metrics["num_parameters"] = int(lmkan_param_count)

    print("\n=== Final comparison ===")
    print(f"PIKAN  -> Params={pikan_metrics['num_parameters']}, MSE={pikan_metrics['mse']:.6e}, MAE={pikan_metrics['mae']:.6e}, RelL2={pikan_metrics['rel_l2']:.6e}, ResidualMSE={pikan_metrics['residual_mse']:.6e}")
    print(f"LM-KAN -> Params={lmkan_metrics['num_parameters']}, MSE={lmkan_metrics['mse']:.6e}, MAE={lmkan_metrics['mae']:.6e}, RelL2={lmkan_metrics['rel_l2']:.6e}, ResidualMSE={lmkan_metrics['residual_mse']:.6e}")

    # ---------- Save checkpoints ----------
    save_checkpoint(
        OUTPUT_DIR / "best_pikan_ac2_compare.pth",
        model_name="PIKAN",
        model_state_dict=pikan_best_sd,
        config={
            "layers": PIKAN_LAYERS,
            "lr": PIKAN_LR,
            "epochs": PIKAN_EPOCHS,
            "v": PIKAN_V,
            "device": str(device),
            "num_parameters": int(pikan_param_count),
        },
        history=pikan_history,
        metrics=pikan_metrics,
    )

    save_checkpoint(
        OUTPUT_DIR / f"best_lmkan_{LMKAN_BASIS}_ac2_compare.pth",
        model_name="LMKAN",
        model_state_dict=lmkan_best_sd,
        config={
            "width": LMKAN_WIDTH,
            "depth": LMKAN_DEPTH,
            "K": LMKAN_K,
            "basis": LMKAN_BASIS,
            "gamma": LMKAN_GAMMA,
            "metric_hidden": LMKAN_METRIC_HIDDEN,
            "lr": LMKAN_LR,
            "steps": LMKAN_STEPS,
            "device": str(device),
            "num_parameters": int(lmkan_param_count),
        },
        history=lmkan_history,
        metrics=lmkan_metrics,
    )

    np.savez(
        OUTPUT_DIR / f"comparison_results_{LMKAN_BASIS}.npz",
        x=x_vals,
        t=t_vals,
        U_numerical=U_num,
        U_pikan=U_pikan,
        U_lmkan=U_lmkan,
        R_pikan=R_pikan,
        R_lmkan=R_lmkan,
    )

    # ---------- Plots ----------
    plot_loss_comparison(pikan_history, lmkan_history, OUTPUT_DIR / f"loss_compare_pikan_vs_lmkan_{LMKAN_BASIS}.png")
    plot_component_losses(pikan_history, lmkan_history, OUTPUT_DIR / f"loss_parts_pikan_vs_lmkan_{LMKAN_BASIS}.png")
    plot_solution_comparison(U_num, U_pikan, U_lmkan, x_vals, t_vals, OUTPUT_DIR / f"solutions_pikan_lmkan_numerical_{LMKAN_BASIS}.png")
    plot_error_and_residual(U_num, U_pikan, U_lmkan, R_pikan, R_lmkan, OUTPUT_DIR / f"errors_residuals_pikan_lmkan_{LMKAN_BASIS}.png")
    plot_time_slices(x_vals, t_vals, U_num, U_pikan, U_lmkan, str(OUTPUT_DIR / f"timeslice_compare_{LMKAN_BASIS}"))

    summary_lines = [
        "Allen-Cahn case2 comparison summary\n",
        f"PIKAN parameter count: {pikan_param_count}\n",
        f"LM-KAN parameter count: {lmkan_param_count}\n",
        f"PIKAN metrics: {pikan_metrics}\n",
        f"LM-KAN metrics: {lmkan_metrics}\n",
        f"LM-KAN basis: {LMKAN_BASIS}\n",
        f"Saved outputs in: {OUTPUT_DIR.resolve()}\n",
    ]
    with open(OUTPUT_DIR / f"summary_{LMKAN_BASIS}.txt", "w", encoding="utf-8") as f:
        f.writelines(summary_lines)

    print(f"\nAll outputs saved to: {OUTPUT_DIR.resolve()}")
```
